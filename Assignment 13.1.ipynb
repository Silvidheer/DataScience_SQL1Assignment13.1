```python
from io import StringIO
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import requests
```

```python
# Read the dataset
adult_df = pd.read_table('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', sep='\s+', header=None)
adult_df.columns = ['age','workclass','fnlwgt','education','education_num','marital_status','occupation','relationship','race',
                    'sex', 'capital-gain','capital-loss','hours-per-week','native-country','class-probability']
```

```python
#2. Show me the average hours per week of all men who are working in private sector
q = '''Select avg(hoursperweek) from adult_df where sex = 'Male,' and workclass = 'Private,' '''
```

```python
#3. Show me the frequency table for education, occupation and relationship, separately
```

```python
education_freq = adult_df.groupby(["education"])
```

```python
education_freq.describe()
```

```python
occupation_frequency = adult_df.groupby("occupation")
occupation_frequency.describe()
```

```python
relationship_frequency = adult_df.groupby("relationship")
relationship_frequency.describe()
```

```python
# 4. Are there any people who are married, working in private sector and having a masters degree
q = '''Select * from adult_df where marital_status like '%Married-%' and workclass = 'Private,' 
    and education = 'Masters,' ''' 
pysqldf(q)
```

```python
# 5. What is the average, minimum and maximum age group for people working in different sectors
q = '''Select workclass, avg(age),
        min(age), max(age)
        from adult_df group by workclass'''
pysqldf(q)
```

```python
# 6. Calculate age distribution by country
q = '''Select nativecountry, 
        case 
            when age <= 19 then 'Under 19'
            when age > 19 and age <= 80 then 'Between 19 and 80'
            when age > 80 then 'Above 80'
        End as age_range,
        count(*) as Count 
        from adult_df
        group by age_range, nativecountry
        order by nativecountry'''
pysqldf(q)
```

```python
# 7. Compute a new column as 'Net-Capital-Gain' from the two columns 'capital-gain' and 'capital-loss'
q = '''Select (capitalgain - capitalloss) as netcapitalgain from adult_df'''
pysqldf(q)
```
